## Opérateur d'assignation

Les objets sont des conteneurs qui permettent de stocker des informations telles qu'une valeur, une phrase, ou un autre morceau de code comme une fonction (nous y reviendrons plus tard). Pour définir ces objets, vous devez utiliser l'opérateur d'assignation.

| Description | Opérateur |
|:-           | :-:      |
| Assignation | **=**    |

L'opérateur d'assignation est le symbole égal **"="** en Python.


```python
niveau_pikachu = 3
niveau_pikachu
```

Remarquez que dans ce cas, nous n'avons pas utilisé la fonction `print()` à la ligne 2 et pourtant le notebook a retourné une valeur. Les notebooks retournent toujours la dernière instruction.

```python
print(pokemon_1, pikachu_type)
print(pokemon_2, charmander_type)
```

```python
diff_hp = charmander_hp - pikachu_hp
diff_hp
```

```python
hp_moyen = (charmander_hp + pikachu_hp) / 2
hp_moyen
```

### Règles de nommage
- Python est sensible à la casse.
- Les noms d'objets doivent commencer par une lettre.
- Les noms d'objets ne peuvent contenir que des lettres, des chiffres et des underscores.
- Les noms d'objets doivent être descriptifs.

*TIP :* Utilisez uniquement des mots en minuscules séparés par `_` lors du nommage de vos objets.
